***
# <font color='red'>Missing Value</font>

```python
# Let us capture all the nan value
# First lets handle categorical featuers which are missing

feature_nan = [feature for feature in dataset.columns if dataset[feature].isnull().sum()>1 and dataset[feature].dtypes == 'O']

for feature in feature_nan:
    print(f"{feature}:    {np.round(dataset[feature].isnull().mean()*100, 2)}% missing values")
```

```python
# Replace missing value with a new label
def replace_cat_feature(dataset, feature_nan):
    data = dataset.copy()
    data[feature_nan] = data[feature_nan].fillna('Missing')
    
    return data

dataset = replace_cat_feature(dataset, feature_nan)
dataset[feature_nan].isnull().sum()
```

```python
# Now lets check for numerical variables the contains missing values
numerical_with_nan = [feature for feature in dataset.columns if dataset[feature].isnull().sum()>1 and dataset[feature].dtypes != 'O']

for feature in numerical_with_nan:
    print(f"{feature}:  {np.round(dataset[feature].isnull().mean()*100, 2)}% missing value")
```

```python
# Replacing the numerical missing values with median
# As we have seen that there are lot of outliers present therefore we'll consider median 

for feature in numerical_with_nan:
    median_value = dataset[feature].median()
    
    # create a new feature to capture naan value
    dataset[feature+'_nan'] = np.where(dataset[feature].isnull(), 1, 0)
    dataset[feature].fillna(median_value, inplace=True)
    
dataset[numerical_with_nan].isnull().sum()
```

```python
dataset.head()
```

```python
# For temporal variable (Date time variables)

for feature in ['YearBuilt', 'YearRemodAdd', 'GarageYrBlt']:
    dataset[feature] = dataset['YrSold'] - dataset[feature]
```

***
# <font color='red'>Numerical Variables</font>
Since the numerical variables are skewed we will perform log normal distribution

```python
dataset.head()
```

```python
import numpy as np

num_features = ['LotFrontage', 'LotArea', '1stFlrSF', 'GrLivArea', 'SalePrice']

for feature in num_features:
    dataset[feature] = np.log(dataset[feature])
```

```python
dataset.head()
```

***
# <font color='red'>Handling Rare Categorical Feature</font>
We will remove categorical variables that are present less than 1% of the observation. Because this category will not provide that much weight to the output.

```python
categorical_features = [feature for feature in dataset.columns if dataset[feature].dtype == 'O']
```

```python
categorical_features
```

```python
for feature in categorical_features:
    temp =dataset.groupby(feature)['SalePrice'].count()/len(dataset)
    # groupby feature will group all the featrue column wrt all the categorical feature 
    # in that column
    temp_df = temp[temp > 0.01].index
    dataset[feature] = np.where(dataset[feature].isin(temp_df), dataset[feature], 'Rare_var')
```

```python
dataset.head()
```

***
# <font color='red'>Feature Scaling</font>
we need to try both & the one which gives the best result is to be considered.

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
feature_scale = [feature for feature in dataset.columns if dataset[feature].dtypes!='O' and feature not in ['Id', 'SalePrice']]

scalar = MinMaxScaler()
scalar.fit(dataset[feature_scale])
```

```python
# transform the train & test set, and add on the Id and SalePrice variables
data = pd.concat([dataset[['Id', 'SalePrice']].reset_index(drop=True), 
                    pd.DataFrame(scalar.transform(dataset[feature_scale]), columns=feature_scale)],
                    axis=1)
```

```python
# Convert this to csv so that we do not have to do it again & again
data.to_csv('X_train.csv',index=False)
```
